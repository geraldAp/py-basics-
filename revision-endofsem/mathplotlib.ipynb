```python
df = pd.read_csv('telco_churn.csv')
```

```python
df.head()
```

```python
x_values = df[['Total eve calls']]
y_values = df[['Total eve minutes']]


# Create the plot 
plt.figure(figsize=(12,7)) # Set the figure size
plt.plot(x_values, y_values, marker='o' , linestyle='-') # Plotting the line with markers 

# Customize the plot 
plt.xlabel('Total eve calls')
plt.ylabel('Total eve minutes')
plt.title('Total eve calls vs Total eve minutes')
plt.grid(True)

plt.show()
```
